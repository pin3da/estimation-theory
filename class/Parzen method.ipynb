```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from numpy import random as rd
from scipy.stats import norm as normal

def p(x): 
    return 0.3 * normal.pdf(x, -2, 0.8) + 0.7 * normal.pdf(x, 2, 0.8)


def draw():
    pi = rd.uniform()
    index = 0 if (pi <= 0.3) else 1
    # index = rd.binomial(1, 0.3)
    mean = [-2, 2]
    return rd.normal(mean[index], 0.8)
    

def esti(N, h, x):
    ans = 0
    den = np.sqrt(2.0 * np.pi * (h ** 2.0))
    for i in range(N):
        xn = draw()
        num = np.exp(- ((x - xn) ** 2.0) / (2.0 * (h ** 2))) 
        ans += num / den
    return ans / N

x = np.linspace(-5, 5, 300)



Nexp = [10, 100, 1000]
Hexp = [2, 0.3, 0.5, 0.4]

for n in Nexp:
    for h in Hexp:
        plt.plot(x, p(x))
        y = []
        for i in x:
            y.append(esti(n, h, i))
        plt.plot(x, y)
        plt.title("Using n = %d, h = %f" % (n, h))
        plt.show()
```
