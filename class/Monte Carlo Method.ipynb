```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from numpy import random as rd

def f2(x):
    return np.sin(x)

def real2(x):
    return -np.cos(x) + np.cos(0)
    
def f(x):
    return x ** 2.0

def mc(N, fun):
    ans = 0
    for i in range(N):
        x = rd.uniform()
        ans += fun(x) / N
    return ans

def real(x):
    return (x ** 3.0) / 3.0

# print("Valor real", real(1))

X = []
Y = []
Yr = []

for n in range(600):
    X.append(n)
    Y.append(mc(n, f))
    Yr.append(real(1))
    # print("Valor aproximado usando %f muestras: %f" % (n, mc(n, f)))
    
plt.plot(X, Y)
plt.plot(X, Yr, 'r')
plt.show()

X = []
Y = []
Yr = []

for n in range(600):
    X.append(n)
    Y.append(mc(n, f2))
    Yr.append(real2(1))
    # print("Valor aproximado usando %f muestras: %f" % (n, mc(n, f)))
plt.plot(X, Y)
plt.plot(X, Yr, 'r')
plt.show()
```
